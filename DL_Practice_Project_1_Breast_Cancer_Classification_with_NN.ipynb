# Breast Cancer Classification with a Simple Neural Network

This project demonstrates a simple neural network model for classifying breast cancer as either malignant or benign based on the Wisconsin Breast Cancer Diagnostic dataset.

## Dataset

The dataset used in this project is the Wisconsin Breast Cancer Diagnostic dataset, which is included in scikit-learn. It contains features computed from digitized images of fine needle aspirates (FNA) of breast masses. The target variable indicates whether the breast mass is malignant (0) or benign (1).

The neural network model consists of:

- A Flatten layer to convert the input data into a 1D array.
- A Dense layer with 20 units and ReLU activation.
- A Dense layer with 2 units and Sigmoid activation for binary classification.

**Importing the Dependencies**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn.datasets
from sklearn.model_selection import train_test_split
```

Data Collection & Processing

```python
# loading the data from sklearn
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
print(breast_cancer_dataset)
```

```python
# loading the data to a data frame
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# print the first 5 rows of the dataframe
data_frame.head()
```

```python
# adding the 'target' column to the data frame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# print last 5 rows of the dataframe
data_frame.tail()
```

```python
# number of rows and columns in the dataset
data_frame.shape
```

```python
# getting some information about the data
data_frame.info()
```

```python
# checking for missing values
data_frame.isnull().sum()
```

```python
# statistical measures about the data
data_frame.describe()
```

0 --> Malignant

```python
data_frame.groupby('label').mean()
```

Separating the features and target

```python
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

Splitting the data into training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=2)
```

```python
print(X.shape, X_train.shape, X_test.shape)
```

Standardize the data

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()

X_train_std = scaler.fit_transform(X_train)

X_test_std = scaler.transform(X_test)
```

**Building the Neural Network**

download.png

```python
# importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of Neural Network

model = keras.Sequential([
                          keras.layers.Flatten(input_shape=(30,)),
                          keras.layers.Dense(20, activation='relu'),
                          keras.layers.Dense(2, activation='sigmoid')
])
```

```python
# compiling the Neural Network

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])
```

```python
# training the Meural Network

history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

Visualizing accuracy and loss

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'upper right')
```

Accuracy of the model on test data

```python
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

**Building the predictive system**

```python
input_data = (11.76,21.6,74.72,427.9,0.08637,0.04966,0.01657,0.01115,0.1495,0.05888,0.4062,1.21,2.635,28.47,0.005857,0.009758,0.01168,0.007445,0.02406,0.001769,12.98,25.72,82.98,516.5,0.1085,0.08615,0.05523,0.03715,0.2433,0.06563)

# change the input_data to a numpy array
input_data_as_numpy_array = np.asarray(input_data)

# reshape the numpy array as we are predicting for one data point
input_data_reshaped = input_data_as_numpy_array.reshape(1,-1)

# standardizing the input data
input_data_std = scaler.transform(input_data_reshaped)

prediction = model.predict(input_data_std)
print(prediction)

prediction_label = [np.argmax(prediction)]
print(prediction_label)

if(prediction_label[0] == 0):
  print('The tumor is Malignant')

else:
  print('The tumor is Benign')
```
